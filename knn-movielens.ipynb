```python
Ratings = ratings.pivot(index = 'user_id', columns ='movie_id', values = 'rating').fillna(0)
Ratings.head()
```

```python
from scipy.sparse import csr_matrix
Rating_matrix = csr_matrix(Ratings.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute',n_neighbors=7)
model_knn.fit(Rating_matrix)
```

```python
#query_index = np.random.choice(Ratings.shape[0])
def neighbours(query_index):
    distances, indices = model_knn.kneighbors(Ratings.iloc[query_index, :].values.reshape(1,-1), n_neighbors = 21)
    k=[]
    for i in range(1, len(distances.flatten())):
        k.append(Ratings.index[indices.flatten()[i]])
    return k
        
```

```python
neighbours(1)
```

```python
def similarity(uid):
    distances, indices = model_knn.kneighbors(Ratings.iloc[uid, :].values.reshape(1, -1), n_neighbors = 11)

    for i in range(1, len(distances.flatten())):
        print('{0}: {1}, have similarity {2}'.format(i, Ratings.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
similarity(1)
```

```python
def moviemeta(movieid):
    title=movies[movies.movie_id==movieid]['title'].to_string(index=False)
    return title
```

```python
def favmovie(userId,N=20):
    userRatings= ratings[ratings['user_id']==userId]
    sortedRatings= pd.DataFrame.sort_values(userRatings,['rating'],ascending=[0])[:N]
    sortedRatings['title']= sortedRatings['movie_id'].apply(moviemeta)
    return sortedRatings
```
