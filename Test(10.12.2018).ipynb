```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
```

```python
data=pd.read_csv('C:/Users/User/Downloads/BikeSharingData.csv')
data.columns
```

### Q.1) Draw scatter plots for casual vs registered users

```python
plt.scatter(data['casual'],data['registered'])
plt.show()
```

### Q.2) Draw a histogram showing bike sharing data for various seasons

```python
g = sns.FacetGrid(data, col='season')
g.map(plt.hist, 'cnt')
plt.show()
```

### Q.3) Draw a histogram for various weather situations.

```python
data['weathersit'].value_counts()
```

```python
plt.hist(data['weathersit'], 10, color = 'pink',  histtype = 'stepfilled', rwidth = 0.8)  
plt.xlabel('Weather Situations') 
plt.ylabel('Count') 
plt.title('Weather Situations')
plt.show() 
```

### Q.5) Draw a bar chart for bike sharing data for each month.

```python
plt.bar(data['mnth'], data['cnt'], align='center', alpha=0.5)

#plt.xticks(data['mnth'], data['cnt'])
plt.ylabel('Count')
plt.xlabel('month')
plt.show()
```

### Q.6) Draw a pie chart for #5.

```python
sums = data.mnth.groupby(data.mnth).sum()
axis('equal');
pie(sums, labels=sums.index);
show()
```

### Q.8) Draw a box plot for bike sharing data, What is your analysis of the box plot.
